```python
def do_op1(x, y, z):
    return x + y - (2 - z)

def do_op2(a, b):
    return a * b + 3

def do_op3(p):
    return p*p*p
```

```python
import numpy as np
```

```python
x = np.random.random(5)
y = np.random.random(5)
z = np.random.random(5)

final_result = []
```

```python
for i in range(0, 10):
    op1 = do_op1(x[i], y[i], z[i])
    op2 = do_op2(op1, y[i])
    op3 = do_op3(op2)
    final_result.append(op3)
```

## Using Dask @delayed

```python
from dask import delayed, compute

@delayed
def do_op1(x, y, z):
    return x + y - (2 - z)

@delayed
def do_op2(a, b):
    return a * b + 3

@delayed
def do_op3(p):
    return p*p*p
```

```python
final_result = []
for i in range(0, x.size):
    op1 = do_op1(x[i], y[i], z[i])
    op2 = do_op2(op1, y[i])
    op3 = do_op3(op2)
    final_result.append(op3)

final_sum = delayed(sum)(final_result)
final_sum #lazy evaluation : Not computed yet
```

```python
final_sum.compute()
```

```python
final_sum.visualize()
```
